## Câu hỏi 6: Liệt kê các trận đấu và thông tin về leage và country tương ứng

Bảng Match có quan hệ với :
 - Bảng Country thông qua khóa country_id
 - Bảng League thông qua khóa League_id
 
Sử dụng câu lệnh join để kết nối dư liệu từ 3 bảng trên.


```python
result_match_league_country = pd.read_sql("""SELECT c.name as Country, l.name as League_Name
                                                , home_team_api_id, away_team_api_id
                                            FROM match m
                                                INNER JOIN League l on m.League_id = l.id
                                                INNER JOIN country c on m.country_id = c.id
                                            ORDER BY Country, League_Name""", conn)
result_match_league_country
```

## Câu hỏi 7: Liệt kê số lượng trận đấu của mỗi giải đấu bao gồm cả tên giải đấu, sắp xếp theo thứ tự giảm dần số trận đấu

```python
result_count_league = pd.read_sql("""SELECT l.name as League_Name
                                                , count(*) as Count_Match
                                            FROM match m
                                                INNER JOIN League l on m.League_id = l.id
                                            GROUP BY l.name
                                            ORDER BY Count_Match DESC""", conn)
result_count_league
```

## Câu hỏi 8: Liệt kê tổng số bàn thắng của đội nhà và đội khách trong mỗi giải đấu
- Sử dụng câu lệnh Group By

```python
result_total_goal = pd.read_sql("""SELECT l.name as League_Name
                                                , sum(home_team_goal) as total_home_team_goal
                                                , sum(away_team_goal) as total_away_team_goal
                                            FROM match m
                                                INNER JOIN League l on m.League_id = l.id
                                            GROUP BY l.name""", conn)
result_total_goal
```

## Câu hỏi 10: Liệt kê 20 đội với số bàn thắng sân nhà cao nhất
- Sử dung câu lênh GROUP BY, LIMIT

```python
result_team = pd.read_sql("""SELECT t.team_long_name, sum(home_team_goal) as total FROM MATCH m
                                INNER JOIN Team t on m.home_team_api_id = t.team_api_id
                            GROUP BY team_long_name
                            ORDER BY total DESC
                            LIMIT 20""", conn)
result_team
```

## Câu hỏi 11: Liệt kê 20 đội có số bàn thắng trên sân khách cao nhất
- Sử dụng câu lênh GROUP BY, LIMIT

```python
result_team = pd.read_sql("""SELECT t.team_long_name, sum(away_team_goal) as total FROM MATCH m
                                INNER JOIN Team t on m.away_team_api_id = t.team_api_id
                            GROUP BY team_long_name
                            ORDER BY total DESC
                            LIMIT 20""", conn)
result_team
```

## Câu hỏi 12: Liệt kê tên các đội bóng và tổng số bàn thắng mỗi đội, sắp xếp  theo số lượng giảm dần

- Tổng số bàn thắng của một đội bao gồm cả số bàn thắng ở sân nhà và sân khách.
- Sử dụng câu lệnh Union  

```python
result_team = pd.read_sql("""SELECT t.team_long_name, m.total_goal FROM team t
                                INNER JOIN 
                                (SELECT team_api_id, sum(goal) AS total_goal FROM
                                        (SELECT home_team_api_id AS team_api_id, sum(home_team_goal) AS goal FROM match
                                        GROUP BY team_api_id
                                        UNION ALL
                                        SELECT away_team_api_id AS team_api_id, sum(away_team_goal) AS goal FROM match
                                        GROUP BY team_api_id)
                                    GROUP BY team_api_id) AS m on t.team_api_id = m.team_api_id
                                ORDER BY m.total_goal DESC""", conn)
result_team
```

## Câu hỏi 13: Liệt kê tên các đội bóng (long name) và tổng số trận đấu đội đó tham gia, sắp xếp theo thứ tự giảm dần.

```python
result_team = pd.read_sql("""SELECT t.team_long_name, m.total_match FROM team t
                                INNER JOIN 
                                (SELECT team_api_id, sum(count_match) AS total_match FROM
                                        (SELECT home_team_api_id AS team_api_id, count(*) AS count_match FROM match
                                        GROUP BY team_api_id
                                        UNION ALL
                                        SELECT away_team_api_id AS team_api_id, count(*) AS count_match FROM match
                                        GROUP BY team_api_id)
                                    GROUP BY team_api_id) AS m on t.team_api_id = m.team_api_id
                                ORDER BY m.total_match DESC""", conn)
result_team
```

## Câu hỏi 14: Liệt kê số trận thắng, thua và hòa của mỗi đội.

- Nếu một đội có số lượng bàn thắng trên sân nhà > số lượng bàn thắng trên sân khách của chính đội đó, thì đội đó sẽ "thắng" trong trận đấu này
- Nếu một đội có số lượng bàn thắng trên sân nhà < số lượng bàn thắng trên sân khách của chính đội đó, thì đội đó sẽ "thua" trong trận đấu này
- Nếu một đội có số lượng bàn thắng trên sân nhà = số lượng bàn thắng trên sân khách của chính đội đó, thì đội đó sẽ "hòa" trong trận đấu này
- Sử dụng câu lệnh "Case When"


```python
result_team = pd.read_sql("""SELECT TEAM_API_ID, SUM(COUNT_WIN) AS TOTAL_WIN, SUM(COUNT_DRAW) AS TOTAL_DRAW
                                , SUM(COUNT_LOSE) AS TOTAL_LOSE
                                FROM
                                        (SELECT HOME_TEAM_API_ID AS TEAM_API_ID, 
                                            CASE 
                                                WHEN HOME_TEAM_GOAL > AWAY_TEAM_GOAL THEN 1
                                                ELSE 0
                                            END AS COUNT_WIN
                                            , CASE
                                                WHEN HOME_TEAM_GOAL = AWAY_TEAM_GOAL THEN 1
                                                ELSE 0
                                            END AS COUNT_DRAW
                                            , CASE
                                                WHEN HOME_TEAM_GOAL < AWAY_TEAM_GOAL THEN 1
                                                ELSE 0
                                            END AS COUNT_LOSE
                                        FROM MATCH
                                        UNION ALL
                                        SELECT AWAY_TEAM_API_ID AS TEAM_API_ID,
                                            CASE 
                                                WHEN AWAY_TEAM_GOAL > HOME_TEAM_GOAL  THEN 1
                                                ELSE 0
                                            END AS COUNT_WIN
                                            , CASE
                                                WHEN AWAY_TEAM_GOAL = HOME_TEAM_GOAL  THEN 1
                                                ELSE 0
                                            END AS COUNT_DRAW
                                            , CASE
                                                WHEN AWAY_TEAM_GOAL < HOME_TEAM_GOAL THEN 1
                                                ELSE 0
                                            END AS COUNT_LOSE
                                         FROM MATCH
                                        ) 
                                    GROUP BY TEAM_API_ID
                            """, conn)
result_team
```

## Câu hỏi 15: Liệt kê 10 đội có số trận thắng nhiều nhất

```python
result_team = pd.read_sql("""SELECT TEAM_API_ID, SUM(COUNT_WIN) AS TOTAL_WIN, SUM(COUNT_DRAW) AS TOTAL_DRAW
                                , SUM(COUNT_LOSE) AS TOTAL_LOSE
                                FROM
                                        (SELECT HOME_TEAM_API_ID AS TEAM_API_ID, 
                                            CASE 
                                                WHEN HOME_TEAM_GOAL > AWAY_TEAM_GOAL THEN 1
                                                ELSE 0
                                            END AS COUNT_WIN
                                            , CASE
                                                WHEN HOME_TEAM_GOAL = AWAY_TEAM_GOAL THEN 1
                                                ELSE 0
                                            END AS COUNT_DRAW
                                            , CASE
                                                WHEN HOME_TEAM_GOAL < AWAY_TEAM_GOAL THEN 1
                                                ELSE 0
                                            END AS COUNT_LOSE
                                        FROM MATCH
                                        UNION ALL
                                        SELECT AWAY_TEAM_API_ID AS TEAM_API_ID,
                                            CASE 
                                                WHEN AWAY_TEAM_GOAL > HOME_TEAM_GOAL  THEN 1
                                                ELSE 0
                                            END AS COUNT_WIN
                                            , CASE
                                                WHEN AWAY_TEAM_GOAL = HOME_TEAM_GOAL  THEN 1
                                                ELSE 0
                                            END AS COUNT_DRAW
                                            , CASE
                                                WHEN AWAY_TEAM_GOAL < HOME_TEAM_GOAL THEN 1
                                                ELSE 0
                                            END AS COUNT_LOSE
                                         FROM MATCH
                                        ) 
                                    GROUP BY TEAM_API_ID
                                    ORDER BY SUM(COUNT_WIN) DESC
                                    LIMIT 10
                            """, conn)
result_team

```

## Câu hỏi 16: Liệt kê tỉ lệ phần trăm về số trận của mỗi giải trên tổng tất cả các trận

```python
result_team = pd.read_sql("""SELECT LEAGUE_ID, TOTAL_PLAY_LEAGUE, TOTAL_PLAY_ALL_LEAGUE, ROUND(TOTAL_PLAY_LEAGUE * 100 / TOTAL_PLAY_ALL_LEAGUE, 2) AS RATIO
                            FROM
                            (
                                 SELECT LEAGUE_ID, TOTAL_PLAY_LEAGUE, 
                                    (SELECT COUNT(*) AS TOTAL_MATCH FROM MATCH) AS TOTAL_PLAY_ALL_LEAGUE
                                FROM
                                    (
                                        SELECT LEAGUE_ID, COUNT(*) AS TOTAL_PLAY_LEAGUE
                                        FROM MATCH
                                        GROUP BY LEAGUE_ID
                                    )
                            )
                            """, conn)
result_team
```
